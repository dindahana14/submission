## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim apa yang menjadi highest demand untuk bike rental?

```python
day_df.groupby(by='season').agg({
    'casual': 'sum',
    'registered': 'sum',
    'cnt': 'sum'
})
```

```python
seasons = ('Spring', 'Summer', 'Fall', 'Winter')
rentals = [471348, 918589, 1061129, 841613]  
colors = ['#72BCD4', '#FF6F61', '#93C572', '#E67F0D'] 
explode = (0.05, 0, 0, 0)  

plt.figure(figsize=(8, 8))
plt.pie(
    x=rentals,
    labels=seasons,
    autopct='%1.1f%%',  
    colors=colors,
    explode=explode,
    shadow=True,  
    startangle=90 
)

# Title for the pie chart
plt.title("Proportion of Total Bike Rentals by Season", fontsize=16)
plt.show()
```

Insights:

- Musim Gugur (Fall) merupakan musim dengan proporsi penyewaan sepeda tertinggi, mencapai 32.2%. Ini mungkin menunjukkan bahwa cuaca pada musim ini lebih mendukung untuk aktivitas luar ruangan seperti bersepeda.

- Musim Panas (Summer) memiliki proporsi terbesar kedua dengan 27.9%. Meski musim ini cenderung panas, tetapi tetap menjadi periode yang populer untuk penyewaan sepeda, mungkin karena waktu liburan atau durasi siang yang lebih lama.

- Musim Dingin (Winter) memiliki proporsi 25.6%, yang relatif tinggi untuk musim dengan cuaca yang cenderung dingin. Ini bisa disebabkan oleh kebutuhan transportasi meskipun cuaca kurang mendukung, atau mungkin lokasi data ini berasal dari daerah yang musim dinginnya tidak terlalu ekstrem.

- Musim Semi (Spring) memiliki proporsi penyewaan sepeda paling kecil, yaitu 14.3%. Ini bisa jadi karena cuaca musim semi yang mungkin tidak sepenuhnya stabil atau adanya hujan.

### Pertanyaan 2: Bagaimana trend dari bike rentals di tahun 2011 dan 2012?

```python
df_2011 = day_df[day_df['yr'] == 0]
df_2012 = day_df[day_df['yr'] == 1]

monthly_report_2011 = df_2011.resample(rule='M', on='dteday').agg({
    'instant': 'nunique',
    'cnt': 'sum'
})
monthly_report_2012 = df_2012.resample(rule='M', on='dteday').agg({
    'instant': 'nunique',
    'cnt': 'sum'
})

monthly_report_2011.index = monthly_report_2011.index.strftime('%B')
monthly_report_2012.index = monthly_report_2012.index.strftime('%B')

monthly_report_2011 = monthly_report_2011.reset_index()
monthly_report_2012 = monthly_report_2012.reset_index()

monthly_report_2011.rename(columns={
    'instant': 'record',
    'cnt': 'total_rentals'
}, inplace=True)
monthly_report_2012.rename(columns={
    'instant': 'record',
    'cnt': 'total_rentals'
}, inplace=True)

plt.figure(figsize=(16, 8))
plt.plot(monthly_report_2011["dteday"], monthly_report_2011["total_rentals"], 
         marker='o', linewidth=2, color="#72BCD4", label="2011")
plt.plot(monthly_report_2012["dteday"], monthly_report_2012["total_rentals"], 
         marker='o', linewidth=2, color="#FF6F61", label="2012")
plt.title("Number of Orders per Month (2011 vs 2012)", loc="center", fontsize=20)
plt.xticks(ticks=range(len(monthly_report_2011["dteday"])), 
           labels=monthly_report_2011["dteday"], fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Rentals", fontsize=12)
plt.grid(True)
plt.legend(title="Year", fontsize=12)
plt.tight_layout()
plt.show()
```

Insights:
- Secara keseluruhan, jumlah penyewaan sepeda pada tahun 2012 lebih tinggi dibandingkan dengan tahun 2011 di setiap bulan. Hal ini menunjukkan adanya peningkatan penggunaan sepeda yang cukup signifikan pada tahun 2012.
- Kedua tahun mengikuti pola yang mirip, di mana penyewaan sepeda mulai meningkat pada bulan Maret hingga mencapai puncaknya di pertengahan tahun (Mei dan Juni), kemudian menurun kembali menjelang akhir tahun.
- Puncak penyewaan sepeda di tahun 2012 terjadi pada bulan Juni dan September, sedangkan di tahun 2011 puncak penyewaan terjadi di bulan Juni.
- Pada kedua tahun, terlihat penurunan signifikan dalam penyewaan sepeda pada bulan-bulan terakhir (Oktober hingga Desember). Ini mungkin dikarenakan faktor cuaca yang lebih dingin atau kondisi yang kurang mendukung untuk bersepeda.

## Pertanyaan 3: Kapan peak hours rental bikes terjadi?

```python
hourly_rentals = hour_df.groupby('hr').agg({
    'cnt': 'sum'
}).reset_index()

# Plotting the hourly bike rentals
plt.figure(figsize=(10, 6))
plt.plot(hourly_rentals['hr'], hourly_rentals['cnt'], marker='o', color='#4C72B0', linewidth=2)
plt.title('Total Bike Rentals by Hour', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Total Rentals', fontsize=12)
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```

**Insight:**
Puncak Pagi dan Sore: Penyewaan sepeda paling tinggi terjadi pada dua waktu puncak utama:

- Sekitar pukul 8 pagi: Ini kemungkinan mencerminkan jam sibuk pagi ketika orang berangkat kerja atau sekolah.
- Sekitar pukul 5 hingga 6 sore: Ini mencerminkan jam sibuk sore ketika orang pulang kerja atau kegiatan lainnya.
- Penurunan di Tengah Hari: Setelah puncak pagi, penyewaan turun tajam menjelang siang (sekitar pukul 10), lalu naik lagi secara bertahap sekitar pukul 12 hingga 1 siang.

- Penggunaan Rendah di Malam Hari: Penyewaan sepeda menurun drastis setelah pukul 8 malam dan mencapai titik terendah setelah tengah malam.
 
- Penggunaan Pagi Buta yang Rendah: Penggunaan sepeda sangat rendah antara pukul 12 malam hingga 5 pagi, yang mungkin disebabkan oleh waktu istirahat dan kurangnya aktivitas pada jam tersebut.
